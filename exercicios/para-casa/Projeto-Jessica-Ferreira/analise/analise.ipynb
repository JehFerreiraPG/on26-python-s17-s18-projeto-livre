```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_microdados_inep = pd.read_csv("microdados_inep_ed_basica_2022.csv", sep= ";", encoding= "latin1")
df_microdados_inep
```

```python
df_microdados_inep.info()
```

```python
columns_drop = ["CO_REGIAO", "CO_UF", "CO_MUNICIPIO", "CO_MESORREGIAO", "CO_MICRORREGIAO", "CO_DISTRITO"]
df_microdados_inep1 = df_microdados_inep.drop(columns_drop, axis=1)
df_microdados_inep1
```

```python
df_microdados_inep1.info()
```

```python
df_microdados_inep1.to_csv("bs_inep_analise.csv", index=False)

# Base final INEP utilizada na análise.
```

```python
qnt_esc_cads_por_est = df_microdados_inep1["NO_UF"].value_counts()
qnt_esc_cads_por_est

# Quantitativo de escolas cadastradas por estado segundo INEP 2022.
```

```python
porct_esc_cads_por_reg = df_microdados_inep1["NO_REGIAO"].value_counts(normalize=True)
porct_esc_cads_por_reg

# Porcetagem do quantitativo por região.
```

```python
df_pe_inep = df_microdados_inep1[df_microdados_inep1["SG_UF"] =="PE"]
df_pe_inep

# Dataframe separado só de Pernambuco para gerar analisar a questão de quantidade de escolas por municipios.
```

```python
df_mesorregiao = df_pe_inep["NO_MESORREGIAO"].value_counts()
df_mesorregiao.plot(kind="barh", edgecolor='black');
plt.xlabel("")
plt.ylabel("")
plt.title("Escolas por mesorregião")

df_mesorregiao

# Região metropolitana e agreste pernambucano com o dobro da quantidade de escolas cadastradas.
```

```python
df_base_pcr2021 = pd.read_csv("base_de_dados_pcr_situacao_alunos_2021.csv")
df_base_pcr2021
```

```python
df_base_pcr2021.info()
```

```python
columns_drop1 = ["AESCOLNUME", "CESCOLRPAE", "MODENS", "COMPENS", "EESCOLRUAE", "ESERMODESC"]
df_base_pcr2021_1 = df_base_pcr2021.drop(columns_drop1, axis=1)
df_base_pcr2021_1.info()
```

```python
df_base_pcr2022 = pd.read_csv("base_de_dados_pcr_situacao_alunos_2022.csv")
df_base_pcr2022
```

```python
df_base_pcr2022.info()
```

```python
columns_drop2 = ["aescolnume", "cescolrpae", "modens", "compens", "eescolruae", "esermodesc"]
df_base_pcr2022_1 = df_base_pcr2022.drop(columns_drop2, axis=1)
df_base_pcr2022_1.info()
```

```python
df_dados_pcr_21_22 = pd.merge(df_base_pcr2021_1, df_base_pcr2022_1, on="_id")
df_dados_pcr_21_22
```

```python
df_dados_pcr_21_22.info()
```

```python
df_dados_pcr_21_22.isnull().sum()
```

```python
df_dados_pcr_21_22.to_csv("bs_pcr_21_22_analise.csv", index=False)

# Base final PCR utilizada na análise.
```

```python
dado_agrupado_nivel_alunos = df_dados_pcr_21_22.groupby("nmodennome")
dado_agrupado_nivel_alunos
```

```python
contagem_alunos_serie = dado_agrupado_nivel_alunos["_id"].count()
contagem_alunos_serie

# Quantidade de alunos matriculados nos anos de 2021 e 2022 nas escolas da prefeitura da cidade do Recife, separados por nível.
```

```python
df_prc_ens_fund = df_dados_pcr_21_22[(df_dados_pcr_21_22["nmodennome"] == "ENSINO FUNDAMENTAL") & (df_dados_pcr_21_22["NMODENNOME"] == "ENSINO FUNDAMENTAL")]
df_prc_ens_fund

# Dataframe separado só com os alunos do ensino fundamental.
```
